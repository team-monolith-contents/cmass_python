```python
# ÏïåÍ≥†Î¶¨Ï¶ò A) Ï†úÎπÑÎΩëÍ∏∞
import time
from random import randint

num = int(input("Ï†ÑÏ≤¥ Ïù∏ÏõêÏàòÎ•º ÏûÖÎ†•Ìï¥ Ï£ºÏÑ∏Ïöî:"))
teamNum = int(input("ÌåÄÏõê Ïà´ÏûêÎ•º ÏûÖÎ†•Ìï¥ Ï£ºÏÑ∏Ïöî:"))

t1 = time.time()
cls = [0 for _ in range(num)]
check = [0 for _ in range(num)]
for i in range(num):
    randNum = randint(0, num-1)
    while(check[randNum]):
        randNum = randint(0, num-1)
    cls[i] = randNum + 1
    check[randNum] = 1
t2 = time.time()

for i in range(0, num, teamNum):
    print(cls[i:i+teamNum])
print(t2-t1)
```

```python
# ÏïåÍ≥†Î¶¨Ï¶ò B) Ï†úÎπÑÎΩëÍ∏∞
import time
from random import randint

num = int(input("Ï†ÑÏ≤¥ Ïù∏ÏõêÏàòÎ•º ÏûÖÎ†•Ìï¥ Ï£ºÏÑ∏Ïöî:"))
teamNum = int(input("ÌåÄÏõê Ïà´ÏûêÎ•º ÏûÖÎ†•Ìï¥ Ï£ºÏÑ∏Ïöî:"))

t1 = time.time()
cls = [i+1 for i in range(num)]
for _ in range(100):
    randA = randint(0, num-1)
    randB = randint(0, num-1)
    cls[randA], cls[randB] = cls[randB], cls[randA]
t2 = time.time()

for i in range(0, num, teamNum):
    print(cls[i:i+teamNum])
print(t2-t1)
```
