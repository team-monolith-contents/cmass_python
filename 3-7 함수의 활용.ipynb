## **🎯 2. 지역 변수와 전역 변수**
**학습 목표 : 변수의 사용 범위는 어떻게 될까?**

파이선에서 변수의 사용 범위는 함수를 기준으로 구분한다. 변수를 함수 내부에서 선언하면   
해당 함수 내부에서만 사용할 수 있는데, 이것을 **지역 변수**라고 한다.   
이와 반대로, 변수를 함수 외부에서 선언하면 프로그램 내의 모든 영역에서 자유롭게 사용할 수 있는데 이것을 **전역변수**라고 한다.

다음은 10과 20의 약수 개수와 약수 개수의 누계를 계산하는 프로그램이다.

```python
#과제3 : 10과 20의 약수 개수와 약수 개수이 누계 계산 프로그램
def divisorCount(n):
    tot=0
    for i in range(1, n+1): # n, tot, i: 지역변수
        if n%i == 0:
            tot += 1
    print("약수 개수: ", tot)
    return tot

tot = 0   # tot: 전역 변수
tot += divisorCount(10)
tot += divisorCount(20)
print("약수 누적 개수: ", tot)
```

  

함수 바깥에서 선언된 tot는 전역 변수이다. divisorCount ( ) 함수 내부에서 선언된   
n, tot, i는 모두 지역 변수이다. 전역 변수의 tot와 지역 변수의 tot는 서로 다른 변수이다.
