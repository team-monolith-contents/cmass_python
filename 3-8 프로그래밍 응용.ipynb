# 3-8 프로그래밍 응용

## **🎯 1. 프로젝트의 이해 (교과서 P.167)**
**학습 목표 : 다양한 학문 분야의 문제를 소프트웨어로 협력하여 해결할 수 있는가?**

다양한 학문 분야의 문제를 해결하고자 주제를 선정하고, 그 주제를 탐구하는 데 적절한 방법을 계획한 후   
문제 해결 방법이나 탐구 방법을 선정하여 수행하는 것을 프로젝트라고 한다.    

프로젝트에 참여하는 학생들은 학습의 전 과정에 걸쳐 스스로 활동하고, 프로젝트 진행 과정에 대한   
자료와 결과물을 포트폴리오 형태로 정리하여 제출한다.


## **🎯 2. 프로젝트 준비**
**학습 목표 : 프로젝트 준비 단계에서 팀 구성과 역할을 명확히 분담할 수 있는가**

복잡한 문제일 경우 혼자만의 힘으로는 해결하기 어렵다. 따라서 다양한 능력을 갖춘 사람들이 하나의 팀을  
구성하고 팀원들과 함께 문제를 협력적으로 해결하는 것은 매우 중요하다. 프로젝트팀은 팀장, 프로그래머,  
기획 및 발표자로 구성할 수 있다.  




## **🎯 3. 프로젝트 수행**
**학습 목표 : 프로젝트 수행 절차와 각 절차에서 주의할 점에 대해서 설명할 수 있는가?**

**1️⃣ 문제 발견**

문제를 해결하기 위해 먼저 해야할 것은 진짜 문제를 인식하는 것이다.   
이때 실제로 문제를 겪고 있는 당사자가 어떤 문제를 겪고 있는지
공감하는 것이 중요하다.

**2️⃣ 문제 정의**

해결하고 싶은 문제를 발견했다면 문제를 새로운 관점에서 정의하고,   
이 문제를 어떻게 해결할 수 있을지 생각하기 위한 핵심 질문을 만든다.   



**3️⃣ 아이디어 구상**

핵심 질문을 만드는 이유는 문제를 해결하기 위한 팀원들의 다양한 아이디어를 발산시키기 위해서이다.   
핵심 질문에 대한 다양한 대답을 브레인스토밍하면서 팀원들의 생각을 듣는다.   

  

팀원들이 제시한 의견 중에서 문제를 해결할 수 있으면서 프로그래밍을 통해 실현 가능한 아이디어를 두세 가지 정한다



**4️⃣ 알고리즘 설계**

결정된 아이디어들을 구현하기 위한 다양한 알고리즘을 알고리즘 카드, 순서도 등으로 표현한다.   
이때, 알고리즘을 프로그램으로 작성 시 고려해야할 요소가 무엇인지 생각한다.   
또한, 다양한 알고리즘을 비교하며 각 알고리즘의 장단점도 분석한다.  


**5️⃣ 프로그래밍**

알고리즘을 프로그램으로 구현하기 위해 필요한 변수를 설계하고 알고리즘을 프로그램으로 구현한다.  
프로그램을 작성하면서 오류가 발생하면 수정하고, 프로그램을 완성한 이후에도 더 나은 프로그램으로   
개선하기 위해 지속적으로 수정·보완한다.     



**6️⃣ 발표 및 평가**  
프로젝트 진행 과정에 대해 팀 단위 평가와 개인 단위 평가를 하고 프로젝트 전체 과정에 대해 발표한다.   
이때 다른 팀의 발표를 바른 자세로 경청하는 태도를 가져야 한다. 또한, 다른 팀의 성공과 실패의   
경험으로부터 우리 팀이 경험하지 못했던 것도 배워야 한다.  




## 함께 해결하기
프로그래밍으로 우리 주변의 문제를 해결하기 위한 프로젝트 팀을 구성하고, 각자의 역할을 체크해 보자.
