## **🎯 2. 변수의 활용**
**학습 목표 : 변수를 사용하여 프로그램을 작성할 수 있을까?**

파이선은 변수를 정의할 때 자료형을 지정하지 않고 변수에 저장되는 값의 종류에 따라 자동으로 자료형이 부여된다.  
num1이라는 변수에 10을 넣으면 num1의 자료형은 정수형이된다. 그 다음에 3.14를 넣으면 실수형으로 변경된다. 

```python
#과제 1 : 숫자의 type 출력하기 
num1 = 10
print(type(num1))
```

```python
#과제 2 : 숫자의 type 출력하기 
num1 = 3.14
print(type(num1))
```

과제의 두 변숫값을 서로 교환해보자. 변수는 한 번에 한 개의 값만을 저장하므로 temp라는 임시 변수를 새로 만들어 두 변숫값을 교환할 수 있다. 

```python
#과제 3 : 문자열 교환하기
str1 = "You need python."
str2 = "Life is too short."
temp = str1
str1 = str2
str2 = temp
print(str1)
print(str2)
```

또한, temp와 같은 추가 변수를 사용하지 않고 두 변숫값을 교환할 수 있다. 

```python
#과제 4 : 변수 교환하기
num1 = 100
num2 = 200
num1, num2 = num2, num1
print(num1, num2)
```
