# 스스로 마무리하기

## **🎯 정리하기**
**이 단원의 핵심 질문을 스스로 해결해 보자.**  



## **🎯 평가하기**

  


  

  



```python
m = [0,31,28,31,30,31,30,31,31,30,31,30,31]
n = int(input("알고 싶은 월을 입력해 주세요: "))
s = int(input("1일의 요일을 입력해 주세요: (일요일:0, 토요일:6): "))
print("<<",n,"월의 달력입니다 >>")
print("  S   M   T   W   T   F   S ")

for i in range(s+m[n]) :
    if i < s :
        print("   ",end=" ")
    else :
        print("%3d"%(i-s+1),end=" ")
    if i % 7 == 6 :
        print( )
```
