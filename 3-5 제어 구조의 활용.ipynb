```python
print("[   ]")
myGrade = 0
sumGrade = 0
termCount = int(input("   (1~6): "))
for i in range(termCount):
    totUnit = 0
    totGrade = 0
    print(i+1, end="")
    subjectCount = int(input("   : "))
    for j in range(subjectCount):
        print(j+1, end="")
        unit = int(input("  : "))
        grade = int(input(" : "))
        totUnit += unit
        totGrade += grade*unit
    termGrade = totGrade/totUnit
    sumGrade += termGrade
myGrade = sumGrade/termCount

option = int(input("  (1., 2.): "))

print("  ", myGrade, ".")
if option == 1:
    if myGrade <= 3:
         myLevel = ""
    elif myGrade <= 6:
        myLevel = ""
    else:
        myLevel = ""
    print("", myLevel, ".")
```
